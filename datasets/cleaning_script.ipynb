```python
import numpy as np

x = np.load("2.5k-ideal-recording.npy")

# throw out -1s in the beginning
y = x[np.where(x != -1)]


import matplotlib.pyplot as plt
plt.figure()
plt.plot(y[0:50])


# pad 4 zeros between each datapoint
padded = np.zeros(y.shape[0]*5)
for i in range(y.shape[0]):
    padded[i*5] = y[i]

plt.figure()
plt.plot(padded[0:200])


np.save("2.5k-cleanedup.npy", padded)
```
